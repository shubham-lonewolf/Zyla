```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=data.drop(['Description'],axis=1)
```

```python
data.head()
```

```python
#changing string format to timestamp
```

```python
data['InvoiceDate'] = pd.to_datetime(data.InvoiceDate, format='%m/%d/%Y %H:%M')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe().round(2)
```

```python
data=data.dropna()
data['CustomerID']=data['CustomerID'].astype('int64')
```

```python
data.info()
```

```python
data.describe()
```

```python
#removing quantity less than zero
data=data[data.Quantity > 0]
```

```python
data.describe()
```

```python
#print(data.head())
data['revenue']=data['Quantity']*data['UnitPrice']
```

```python
data.head()
```

```python
data.insert(loc=3, column='month', value=data.InvoiceDate.dt.month)
# +1 to make Monday=1.....until Sunday=7
data.insert(loc=4, column='day', value=(data.InvoiceDate.dt.dayofweek)+1)
data.insert(loc=3,column='year',value=data.InvoiceDate.dt.year)
```

```python
data.head()
```

```python
data['year_month']=data['year']*100+data['month']
data.head()
```

```python
data.groupby(by='year_month')['InvoiceNo'].count()
```

```python
#total order made by customers

data_country_sales=pd.DataFrame(data=data.groupby(by=['Country'])['InvoiceNo'].count().sort_values(ascending=False))
data_country_sales=data_country_sales.reset_index()
```

```python
data_country_sales.head()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x='Country',y='InvoiceNo',data=data_country_sales,palette='Greens_r')
plt.yscale("log")
plt.xticks(rotation=90);
```

```python
data_country_revenue=pd.DataFrame(data.groupby('Country')['revenue'].sum().sort_values(ascending=False))
data_country_revenue=data_country_revenue.reset_index()
data_country_revenue.head()
```

# Insight 1: Top 3 Sales generating countries are UK, Germany and France

```python
plt.figure(figsize=(20,5))
sns.barplot(x='Country',y='revenue',data=data_country_revenue,palette='Greens_r')
plt.yscale("log")
plt.xticks(rotation=90);
```

```python
analysis_data=data.drop('revenue',axis=1)
```

```python
analysis_data
```

```python
X=analysis_data[['StockCode','year','month','day','Country','UnitPrice','year_month']]
#X
```

```python
X=pd.get_dummies(X, dummy_na=True)
X.head()
```

```python
Y=pd.DataFrame(analysis_data['Quantity'])
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
pd.DataFrame(y_pred)
```

```python
y_test
```

```python
model.score(X, Y)
```
